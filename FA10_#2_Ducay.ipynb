The mean of arandom variable X is given by:

E(X) = ∑x P(X = x)

For a binomial distribution, we have the formula given as:

P(X = x) = (n choose x) p^x (1-p)^(n-x)

Calculating the mean:

E(X) = ∑x (n choose x) p^x (1-p)^(n-x)

By the binomial theorem:

(p + q)^n = ∑x (n choose x) p^x q^(n-x)

note that q = 1-p

We differentiate left and right equations with respect to p:

n(p + q)^(n-1) = ∑x (n choose x) x p^(x-1) q^(n-x)

Multiplying by p:

np(p + q)^(n-1) = ∑x (n choose x) x p^x q^(n-x)

We let p + q = 1:

np = ∑x (n choose x) x p^x q^(n-x)

Therefore:

E(X) = ∑x P(X = x) = np

Q.E.D.



The variance of a random variable X is given by the formula:

Var(X) = E(X^2) - (E(X))^2

To calculate E(X^2), we have:

E(X^2) = ∑x x^2 P(X = x)

By the binomial distribution formula:

E(X^2) = ∑x x^2 (n choose x) p^x (1-p)^(n-x)

By expansion:

E(X^2) = ∑x x(x-1) (n choose x) p^x (1-p)^(n-x) + ∑x x (n choose x) p^x (1-p)^(n-x)

Simplifying the first sum:

∑x x(x-1) (n choose x) p^x (1-p)^(n-x) = n(n-1)p^2 ∑x (n-2 choose x-2) p^(x-2) (1-p)^(n-x)

By the binomial theorem:

∑x (n-2 choose x-2) p^(x-2) (1-p)^(n-x) = (p + q)^(n-2) = 1

∑x x(x-1) (n choose x) p^x (1-p)^(n-x) = n(n-1)p^2

Simplifying the second sum:

∑x x (n choose x) p^x (1-p)^(n-x) = np

Therefore, we get:

E(X^2) = np + n(n-1)p^2

We know that:

Var(X) = E(X^2) - (E(X))^2

By substitution: 

Var(X) = np + n(n-1)p^2 - (np)^2
