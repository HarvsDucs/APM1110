```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# Get the BTC returns data
returns = merged['returns'].dropna()

# Plot the histogram of BTC returns
sns.histplot(data=merged, x='returns', bins=1000,)
plt.xlim([-250, 250])
plt.ylim([0, 150000])

# Calculate the mean and standard deviation of BTC returns
mu, std = stats.norm.fit(returns)

# Generate x values for the normal distribution curve
x = np.linspace(-250, 250, 1000)

# Calculate the y values for the normal distribution curve
y = stats.norm.pdf(x, loc=mu, scale=std) * len(returns) * (x[1] - x[0])

# Plot the normal distribution curve on top of the histogram
plt.plot(x, y, 'r--', linewidth=2)
```

The normal distribution was almost alike the btc returns distribution, but surely there are other distributions that better fit the BTC 1 minute returns distribution data.

```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# Get the BTC returns data
returns = merged['returns'].dropna()

# Plot the histogram of BTC returns
sns.histplot(data=merged, x='returns', bins=1000,)
plt.xlim([-250, 250])
plt.ylim([0, 150000])

# Calculate the mean and scale parameter of the Laplace distribution for BTC returns
mu, b = stats.laplace.fit(returns)

# Generate x values for the Laplace distribution curve
x = np.linspace(-250, 250, 1000)

# Calculate the y values for the Laplace distribution curve
y = stats.laplace.pdf(x, loc=mu, scale=b) * len(returns) * (x[1] - x[0])

# Plot the Laplace distribution curve on top of the histogram
plt.plot(x, y, 'g--', linewidth=2)
```

As we can see, the laplace distribution fits BTC 1 minute returns distribution the most among the distrubutions we tried to fit. The laplace distribution was similar to btc returns distribution data in terms of its peak right in the middle.

### Testing for Shapiro-Wilk normality test

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import shapiro

# Create a histogram using seaborn
sns.histplot(data=merged, x='returns', bins=1000,)
plt.xlim([-25, 25])

# Performing Shapiro-Wilk test
statistic, pvalue = shapiro(merged['returns'])

# Print test results
if pvalue < 0.05:
    print("The data is not normally distributed (reject H0)")
else:
    print("The data is normally distributed (fail to reject H0)")
print(f"Statistic: {statistic:.4f}, p-value: {pvalue:.4f}")
```
